```python
s1 = set()
s2 = set()

N1 = eval(input("Enter the number of elements in the first element in 1st set \n"))
N2 = eval(input("Enter the number of elements in the second element in 2nd set \n"))

for i in range(N1):
    x = int(input("Enter the value at: "+ str(i) + " "+ "1st set\n"))
    s1.add(x)
    
for i in range(N2):
    y = int(input("Enter the value at: "+ str(i)+" " + "2st set\n"))
    s2.add(y)
    
#display the number of elements for each set
print("Elements in 1st set is: ", len(s1))
print("Elements in 2nd set is: ", len(s2))

#display the elements in first set
print("Elements in the 1st set is: ", s1)

#display the elements in second set
print("Elements in the 1st set is: ", s2)

#display the elements that are common in both sets
print("Elements that are common: ", s1.intersection(s2))

#display all elements in both sets
print("Both sets are: ",s1.union(s2))
```
