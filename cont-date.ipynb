```python
import datetime

y = eval(input("Enter the year: \n")) 
m = eval(input("Enter the month: \n"))
d = eval(input("Enter the date: \n"))
h = eval(input("Enter the hours: \n"))
m = eval(input("Enter the minutes: \n")) 
s = eval(input("Enter the second: \n"))
ms = eval(input("Enter the microsecond: \n"))
dto = datetime.datetime(y,m,d,h,m,s,ms)
print("The entered date is:", dto)

#extract the year
year = dto.year
print("The year is : ",year)
```

```python
from datetime import date
 
def numOfDays(date1, date2):
    return (date2-date1).days
     
# Difference between dates
date1 = date(2020, 12, 13)
date2 = date(2022, 11, 4)
print(numOfDays(date1, date2), "days")
```
