```python
k = list()
v = list()
d = dict()

N = eval(input("Input the number of students: \n"))

for i in range(0, N):
    x= input("Enter a key at: ", str(i)  +"\n")
    k.append(x)
    
for i in range(0, N):
    y= input("Enter a value at: ", str(i)  +"\n")
    v.append(y)
    
d= dict(zip(k,v))

print("student no",d.keys())
print("student no",d.values())

print("Reg/No \t Fullname")
for k,v in d.items():
    print(k, "\t",v)
    
```
