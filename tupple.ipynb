```python
t =tuple()
N = int(input('Enter the population size: \n'))

for i in range (N):
    x = eval(input("Enter the score at: "+str(i)+"\n"))
    t = t+(x,)
    #display the total score
print('Total score = ', sum(t)) 

#display the number of score
print('Population size = ', len(t))
#display the min and max
print('Highest = ',max(t), 'lowest = ',min(t))
#display the average
print('Average = ',sum(t)/len(t))
#Variance of the score

z=0
u = sum(t)/len(t)
for i in range (len(t)):
    z = (z+(t[i]-u)) **2 
    
    var = z/N
    sd = (var)**(1/2)
print("Variance is = ", z)
print("Standard deviation = ",sd)

    
```
