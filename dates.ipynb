```python
from datetime import date
d = int(input("Enter the date: \n"))
m = int(input("Enter the month: \n"))
y = int(input("Enter the year: \n"))

dt = date(y,m,d) #using date() constructor
print("The date is: ", dt)
```

```python
import datetime
hrs = int(input("Enter the hours"))
mins = int(input("Enter the minutes"))
secs = int(input("Enter the seconds"))
secm = int(input("Enter the microseconds"))
to = datetime.time(hour=hrs,minute=mins,second=secs,microsecond=secm)
print("The time is: ",to)
```
